```
X=np.array(b[l[0:len(l)-1]])
Y=np.array(a['target'])
```

```
#here negative class has 90% of data and positive class has 10% of total data so if we train our model using this data it will converge to negative value and our answer will always zero
#So we have to take training data in almost equal
#Here i have taken this ratio to 1:5(positive:negative)
```

```
xtrain1=[]
ytrain1=[]
xtrain2=[]
ytrain2=[]
for i in range(Y.shape[0]):
  if (Y[i]==0):
    if (len(xtrain1)<2000):
      xtrain1.append(X[i])
      ytrain1.append(0)
  else:
    xtrain2.append(X[i])
    ytrain2.append(1)
```

```
import tensorflow as tf
```

```
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization
from keras.callbacks import TensorBoard 
from keras.preprocessing.image import ImageDataGenerator
```

```
import keras
from sklearn.utils import shuffle
```

```
xtrain=np.array(xtrain1+xtrain2)
ytrain=np.array(ytrain1+ytrain2)
xtrain,ytrain=shuffle(xtrain,ytrain,random_state=0)
```

```
#devide into val and training set
xval=xtrain[0:500]
yval=ytrain[0:500]
xtrain=xtrain[500:]
ytrain=ytrain[500:]
```

```
number_cat=2
ytrain = keras.utils.to_categorical(ytrain, number_cat)
yval = keras.utils.to_categorical(yval, number_cat)
```

```
xtrain.shape,ytrain.shape,xval.shape,yval.shape
```

```
model = Sequential()
model.add(Dense(15,activation='relu',input_shape=[9]))

model.add(Dense(15,activation='relu'))

model.add(Dense(2, activation='softmax'))
```

```
model.summary()
```

```
model.compile(loss = "categorical_crossentropy", optimizer= keras.optimizers.Adam(learning_rate=0.0001), metrics = ["accuracy"])
history = model.fit(xtrain,ytrain, batch_size=128,epochs = 500,validation_data=(xval,yval) ,verbose = 1)
```

```
y=model.predict_classes(X)
```
